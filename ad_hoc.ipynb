Вопрос №2
Какова разница между максимальной и минимальной средней ценой продуктов категории "Ноутбуки" между всеми производителями? Ответ округлите до целых.

```python
df = pd.read_csv('/content/products_data.csv')
df1 = pd.read_csv('/content/sales_data.csv')

itog_df = df.merge(df1, left_on='product_id', right_on='product_id')

df_laptop = itog_df.query('category == "Ноутбуки"')

manufacturer_avg_price = df_laptop.groupby('manufacturer')['price_per_unit'].mean()

round(max(manufacturer_avg_price) - min(manufacturer_avg_price))
```

Вопрос №4
Какова доля продаж продуктов категории "Планшеты" от общей выручки за все время (учитывая только продукты, информация о которых есть в products_data)? Ответ округлите до 2 знака после запятой. Разделитель разряда - точка.

```python
df_product = pd.read_csv('/content/products_data.csv')
df_sales = pd.read_csv('/content/sales_data.csv')

join_df = df_product.merge(df_sales, left_on='product_id', right_on='product_id', how='left')

res = join_df[join_df['category'] == 'Планшеты']['total_price'].sum().tolist()

revenue_laptop = (res / join_df['total_price'].sum()).round(2)

revenue_laptop
```

Вопрос №9
Какой производитель продал продукты на наибольшую сумму в регионе "Юг" за март 2023 года?

```python
df_product = pd.read_csv('/content/products_data.csv')
df_sales = pd.read_csv('/content/sales_data.csv')

join_df = df_product.merge(df_sales, left_on='product_id', right_on='product_id')

join_df['order_date'] = pd.to_datetime(join_df['order_date'])

manufacturer_largest_price = join_df[(join_df['region'] == 'Юг') & ((join_df['order_date'] >= '2023-03-01') & (join_df['order_date'] <= '2023-03-31'))]

manufacturer_largest_price.groupby('manufacturer')['total_price'].agg('sum').sort_values(ascending=False)
```
